```python
# Upload your "data.npy" file into your Google Drive.  
# Verify that it is there by printing out its values. 
# Change the directory as you see fit

data = np.load('/content/drive/My Drive/data.npy') 
print(data.shape)
```

```python
# Making blue channel zero
R_channel = data*1
G_channel = data*1
B_channel = data*0

# Making an RGB image using concatenate function 
data_new = np.concatenate((R_channel[..., None], G_channel[..., None], B_channel[..., None]), axis=2)
print(data_new.shape)
```

```python
#Using a mean filter of size 3x3
mean_filter = np.array([[1/9, 1/9, 1/9], [1/9, 1/9, 1/9], [1/9, 1/9, 1/9]], dtype='float')

#Applying the filter onto the data file provided
filter = signal.convolve2d(data,mean_filter,mode='same')

# Making blue channel zero
filter_R_channel = filter*1
filter_G_channel = filter*1
filter_B_channel = filter*0

# Making an RGB image using concatenate function 
Result2 = np.concatenate((filter_R_channel[..., None], filter_G_channel[..., None], filter_B_channel[..., None]), axis=2)
print(Result2.shape)
```

```python
# As for Assignment 1, you can plot the data using matplotlib's imshow() function.
print("Raw Image")
plt.imshow(data, origin="lower") 
plt.show()

#Save and Download the image
print("Image after change in colour space")
plt.imshow(data_new, origin="lower") 
plt.savefig('Result1.png')
files.download("Result1.png")
plt.show()

#Save and Download the image
print("Filtered Image")
plt.imshow(Result2, origin="lower") 
plt.savefig('Result2.png')
files.download("Result2.png")
plt.show()
```
